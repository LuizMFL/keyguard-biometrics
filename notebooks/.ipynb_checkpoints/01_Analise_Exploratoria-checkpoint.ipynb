```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Configuração de estilo visual para gráficos mais profissionais e limpos
sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)

print("Bibliotecas importadas com sucesso!")
```

```python
# Carregando o dataset (ajuste o caminho do arquivo conforme sua estrutura)
df = pd.read_csv('../data/DSL-StrongPasswordData.csv')

# Inspeção inicial
display(df.head())
print(f"\nDimensões do Dataset: {df.shape}")
print(f"Usuários únicos na base: {df['subject'].nunique()}")

# Verificando se há dados nulos que quebrariam a rede neural
print(f"\nValores Nulos:\n{df.isnull().sum().sum()}")
```

```python
# Vamos comparar o usuário 's002' com o usuário 's032'
user_A = df[df['subject'] == 's002']
user_B = df[df['subject'] == 's032']

# Vamos analisar o tempo de voo (Up-Down) entre as teclas 't' e 'i'
feature_analisada = 'UD.t.i'

plt.figure(figsize=(10, 6))
sns.kdeplot(user_A[feature_analisada], label='Usuário A (s002)', fill=True, color='teal')
sns.kdeplot(user_B[feature_analisada], label='Usuário B (s032)', fill=True, color='coral')

plt.title(f"Distribuição do Tempo de Voo entre 't' e 'i' ({feature_analisada})", fontsize=14)
plt.xlabel("Tempo em segundos", fontsize=12)
plt.ylabel("Densidade", fontsize=12)
plt.legend()
plt.show()
```

```python
# Isolando apenas as colunas numéricas de tempo de digitação
features_tempo = df.drop(['subject', 'sessionIndex', 'rep'], axis=1)

# Visualizando os outliers do usuário A através de um Boxplot
plt.figure(figsize=(14, 6))
sns.boxplot(data=user_A[features_tempo.columns[:10]], orient='h', palette='viridis')
plt.title("Boxplot dos 10 primeiros atributos (Detecção de Outliers - Usuário A)")
plt.xlabel("Tempo (segundos)")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# X = Matriz de tempos de digitação (nossas features)
X = df.drop(['subject', 'sessionIndex', 'rep'], axis=1)

# y = Os rótulos dos usuários
y = df['subject']

# Normalização (Z-Score): Crucial para redes baseadas em gradiente!
# Trazemos a média para 0 e o desvio padrão para 1.
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Criando um DataFrame normalizado para exportação
df_final = pd.DataFrame(X_scaled, columns=X.columns)
df_final['subject'] = y.values

# Separando: 80% para Treino (onde a rede aprende) e 20% para Validação/Teste
# O stratify=y garante que teremos a mesma proporção de cada usuário em ambas as bases
df_train, df_test = train_test_split(df_final, test_size=0.20, random_state=42, stratify=df_final['subject'])

print(f"Tamanho do Dataset de Treino: {df_train.shape}")
print(f"Tamanho do Dataset de Teste: {df_test.shape}")

# Exportando as bases tratadas para os próximos notebooks
df_train.to_csv('../data/train_keystroke.csv', index=False)
df_test.to_csv('../data/test_keystroke.csv', index=False)
print("\nDatasets prontos e exportados para a fase de Deep Learning!")
```
